### Infatti la trappola c'Ã¨: La colonna "Total changes" si presenta come numeri ma nei fatti Ã¨ in fomrato object. Se le altre colonne vanno convertite in numeri,
### qui bisogna "castare" il tipo a float perchÃ¨ voglio mantenere il valore originale.

```python
df["TotalCharges"]=df["TotalCharges"].apply(lambda x: float(x) if len(x)>1 else np.nan)
```

```python
from sklearn.preprocessing import LabelEncoder
df.apply(LabelEncoder().fit_transform)
```

### Ora invece vediamo se ci sono caselle vuote nel dataset (aka valori nulli)

```python
df.isna().sum()
```

### Sempre TotalCharges... Ma questi valori sono nulli per qualche motivo o sono solo dati antati persi? Vediamo...
#### TotalCharges Ã¨ l'addebito totale, e dunque Ã¨ legato ai valori di tenure(tempo di sottoscrizione) e MonthlyCharges(addebito mensile)

```python
df.loc[df["TotalCharges"].isna(),["tenure","MonthlyCharges", "TotalCharges"]]
```

#### Il TotalCharges Ã¨ nullo quando il tenure Ã¨ 0 ovvero quando il cliente Ã¨ stato per 0 mesi, appunto, un cliente. Il valore nullo puo essere allora sostituito da 0 in quanto non vi Ã¨ stato nessun addebito

```python
df["TotalCharges"].fillna(0,inplace=True)
df["TotalCharges"]
```

### Sistemati i valori, posso procedere a convertire i valoroi non numerici in numeri MA facendo occhio a tralasciare le colonne che non vanno modificate:

```python
from sklearn.preprocessing import LabelEncoder

for label in df.columns:
    if(label not in ['TotalCharges', 'MonthlyCharges']):   #applica la trnsform sse non ci si trova sulle colonne in fomrato float (le colonne in int non vengono cambiate dalla funzione)
        df[label] = LabelEncoder().fit_transform(df[label])
df.head()   #per vedere se funziona 
```

### E ora divido il dataset in piu file

```python
from sklearn.model_selection import train_test_split
#split the data into train and test set
train,test = train_test_split(pd.read_csv('Telco-Customer-Churn.partial.csv', sep=','), test_size=0.23, random_state=42) #Prendo il file iniziale perchÃ¨ su df ci ho lavorato fino ad ora
#
train.to_csv('train.csv',index=False)
test.to_csv('test.csv',index=False)
```
